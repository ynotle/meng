To solve this problem, we will code an implementation of the cubic spline. To start, we will need

(c) The system of linear equations can be recast as a linear algebra problem as follows:
$[A]\{\text{spline coefficients}\}=\{b\}$. Now define a python function that accepts x and y
data, and returns a numpy array for $[A]$ with the equation coefficients (dependent on the
x’s) and a numpy array for $\{b\}$ with the equation constants (dependent on y’s). $[A]$ should
be 4n x 4n in dimension and $\{b\}$ be 4n x 1 in dimension for a dataset consisting of n+1
data points.

```python
print(len(time))
print(len(temperature))
print(len(heatflow))
```

```python
#we need to create blank matrices first


# each column represnts the different coefficients, so coefficents from left and right should be equal, as derivatives are equal
#thus, we want to set the 4 columns (i) = to the 4 columns (i+1)

#each row represents a new condition we need to meet. 
def matrix_assembly(x_data,y_data):
    #n i s number of segments
    n = len(x_data)-1
    A = np.zeros((4*n,4*n))
    b = np.zeros((4*n,1))
#function values must be equal at intereior points
    row = 0 
    for i in range(0,n):
        A[row,4*i] = x_data[i]**3
        A[row,(4*i)+1] = x_data[i]**2
        A[row,(4*i)+2] = x_data[i]**1
        A[row,(4*i)+3] = 1

        b[row] = y_data[i]

        row += 1
        A[row,4*i] = x_data[i+1]**3
        A[row,(4*i)+1] = x_data[i+1]**2
        A[row,(4*i)+2] = x_data[i+1]**1
        A[row,(4*i)+3] = 1

        b[row] = y_data[i+1]
        row += 1

#first and last values must pass through endpoints (2)
#because first values have the first four columns, then first point you put in 0:4

        #instead of creating four functions per col, we can do a range of columns

    # A[row, 0:4] = [x_data[0]**3, x_data[0]**2,x_data[0]**1,1]
    # b[row] = y_data[0]
    # row += 1

    # #last values
    # A[row,4*n-4:4*n] = [x_data[n]**3, x_data[n]**2,x_data[n]**1,1]  
    # b[row] = y_data[n]
    # row += 1

#condition 3: first derivs @ inner knots must be equal (n-1)

#each two derivaitves require one point, so n-2. 
    for i in range(0,n-1):
        A[row,4*i:4*i + 4] = [3*x_data[i+1]**2,2*x_data[i+1],1,0]

#here, we need to make sure that the derivaitves at each point are equal. 
#derivative - deriavtive = 0, as they *should* be the same. 
#however, this means the derivative from the other directoin (right), needs to be negative.
        A[row,4*(i+1):4*(i+1)+4] = [-3*x_data[i+1]**2, -2*x_data[i+1],-1,0]
        b[row] = 0
        row += 1


#condition 4: 2nd derivs must be equal (n-1)
#kind of copy the function for the first derivative, but take the 2nd one. 
    for i in range(0,n-1):
        A[row,4*i:4*i + 4] = [6*x_data[i+1],2,0,0]
        A[row,4*(i+1):4*(i+1)+4] = [-6*x_data[i+1], -2,0,0]
        b[row] = 0
        row += 1


#condition 5: second derivates at the end points are zero (2)
    A[row, 0:4] = [6*x_data[0],2,0,0]
    b[row] = 0

    row += 1
    #last values
    A[row,4*n-4:4*n] = [6*x_data[n],2,0,0]
    b[row] = 0
    row += 1

    return A, b
```

Build a function `trapezoid()` that accepts the following arguments: the lower bound for integration `a`, the upper bound for integration `b`, and the number of subintervals `n`. If you would like a hint, check out the multiple-segment pseudocode on page 613.

Now compute the integral above using $n = 18$ subintervals. Report your value to four decimal places.

```python
C0=0.08
tao = 2.5
lamda = 0.6
L=30
x_range = np.linspace(0.01,1.5,250, endpoint=True)
t=6
def oxy_cons(C0, t,tao,r,lamda):
  return (C0*(1-np.exp(-t/tao)) * np.exp(-(r/lamda)) )* 2*np.pi*r*L

def trapezoid(a, b, n):
  step = ((b - a) / n)
  sum = 0 
  for i in range(1,n):

    #we want to evaluate oxygen consumption at specific radii, starting at the lower bound and jumping up by a set step number (defined by the limits)

    sum += oxy_cons(C0, t,tao,(a+i*step),lamda) 
  
  integral = ((b-a) * ((oxy_cons(C0, t,tao,a,lamda)) + 2*sum + oxy_cons(C0, t,tao,b,lamda))) / (2*n)

  return integral

trap_vol_oxy = trapezoid(0.01,1.5,18)

print(f"{trap_vol_oxy:.4g}")

```

(b) Given the simplicity of your algorithm, the trapezoidal approximation is great for a first estimate. But you would now like to be more accurate and would like to build the following functions:

`simpsons_one_third()` and `simpsons_three_eighth()` to compute the same integral as in part (a). These two functions take in the same parameters as `trapezoid()`, a lower bound `a`, an upper bound `b`, and the number of subintervals `n`.

Once you are done, compute the above integral numerically, using $n = 18$ for the 1/3rd Simpson's and $n = 18$ for the 3/8th Simpson's. Remember that due to the nature of the two methods, $n$ must be even when using the Simpson's 1/3rd rule and $n$ must be a multiple of 3 for the Simpson's 3/8th rule; fortunately that requirement is met for $n = 18$.

Please report the final integral value of both functions to four decimal places.

```python
def simpsons_one_third(a, b, n):
  step = ((b - a) / n)
  sum_even = 0 
  sum_odd = 0
  for i in range(1,n,2):
    #we want to evaluate oxygen consumption at specific radii, starting at the lower bound and jumping up by a set step number (defined by the limits)

    sum_odd += oxy_cons(C0, t,tao,(a+i*step),lamda) 

  for i in range(2,n-1,2):
    sum_even +=oxy_cons(C0, t,tao,(a+i*step),lamda) 
  fx0=oxy_cons(C0, t,tao,(a),lamda)
  fxn=oxy_cons(C0, t,tao,(b),lamda) 
 
  
  integral = ((b-a) * (fx0 + 4*sum_odd + 2*sum_even + fxn)) / (3*n)
  return integral

def simpsons_three_eighth(a, b, n):
  step = ((b - a) / n)
  fx0=oxy_cons(C0, t,tao,(n),lamda)
  fx1=oxy_cons(C0, t,tao,(n-step),lamda)
  fx2=oxy_cons(C0, t,tao,(n-(2*step)),lamda)
  fx3=oxy_cons(C0, t,tao,(n-(3*step)),lamda)
  integral = ((b-a) * (fx0 + 3*fx1 + 3*fx2 + fx3)) / 8
  return integral

def simpson_int(a,b,n):
  step = ((b-a) / n)
  sum = 0 
  def f(n):
    return oxy_cons(C0, t,tao,(n),lamda)
  if n ==1:
    sum = trapezoid(step,f(n-1),f(n))
  else:
    m = n 
    odd = n % 2 +1
    if odd > 0 and n >1:
      sum = sum + simpsons_three_eighth(a,b,n)
      m=n-3
  if m>1:
    sum = sum + simpsons_one_third(a,b,n)
  return sum


simp_vol_oxy = simpson_int(0.01,1.5,18)
print(f"{simp_vol_oxy:.4g}")
```
