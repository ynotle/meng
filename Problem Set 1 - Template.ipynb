#### Exercise 2:  Identifying indices in arrays 

Write an efficient function that returns the indices for elements in a NumPy array (either 1D or 2D) that are greater than a user-defined value (provided as an input to the function).

````python
#function for comparison
def indice_func(A,limit):
    loc=[]
    if A.ndim==1: #1D function
        for i in range(A.shape[0]): #only loops through "horizontally"
            if A[i] > limit:
                loc.append((i))
    else: #for 2D, loops over both row and column
        for i in range(A.shape[0]):#iterates over row
            for j in range(A.shape[1]):#iterates over col
                if A[i,j]>limit:
                    loc.append((i,j))
                    #only appends values greater than limit
    return loc

#limit is the user defined value
limit=6
#A is array
A = np.arange(12).reshape(3,4)
print(indice_func(A,limit))
````

````python
#function returns the 4 derivatives of g(x)
#based off of the order "n"
def g_deriv(x, n):
    if n % 4 == 0 :
        return np.sin(x) + 3**n*np.cos(3*x)
    elif n % 4 == 1:
        return np.cos(x) - 3**n*np.sin(3*x)
    elif n % 4 == 2 :
        return -np.sin(x) - 3**n*np.cos(3*x)
    elif n%4 == 3:
        return -np.cos(x) + 3**n*np.sin(3*x)

#function is the form of summation taylor series     
def taylor_g(x, a, nth_order):
    appx = 0
    for i in range(nth_order+1):
        appx += ((g_deriv(a,i))/math.factorial(i))*((x-a)**i)
    return appx

#this is the x axis
X=np.linspace(-0.5,1,400, endpoint=True)

plt.figure(figsize=(8, 5), dpi=120)
#all 5 versions of the polynomial
#this is storing their values in list
g0=taylor_g(X,0.3,0)
g1=taylor_g(X,0.3,1)
g2=taylor_g(X,0.3,2)
g3=taylor_g(X,0.3,3)
g4=taylor_g(X,0.3,4)
#plotting respective values
plt.plot(X,g(X),label = "g(x)")
plt.plot(X,g1,label="n=1")
plt.plot(X,g2, label="n=2")
plt.plot(X,g3,label="n=3")
plt.plot(X,g4,label="n=4")

#plot details
plt.legend(title="Legend")
plt.ylim(-2,2)
plt.title("Plot of g(x) and Taylor Approximations")
plt.show()
````

## Problem 5  –  Taylor series and true absolute error

(a) This Taylor series approximation for $e^{-x^2}$ is centered at which point?  

ANSWER: It is centered around zero. This is because it is a Maclaurin series. In the Taylor series summation formula, there is a term (x-a)^n, where a is the center point of the function. However, in e^-x^2, this term is merely x^2, and there is no a term, meaning that a = 0 and that this approximation is centered around the point x=0.

(b) Write a function `taylor_gaussian(x, n)` which calculates the Taylor polynomial approximation for $e^{-x^2}$ by adding together $n+1$ consecutive terms in the summation form (*i.e.*, summing from $i=0$ to $n$) and evaluating at the specified $x$ value. 

Use the `math.factorial` function in Python to calculate factorials.  If you are unsure how to use this function, refer to the official Python documentation at the following link: https://docs.python.org/3/library/math.html#math.factorial.

````python
#function for the e^-x^2 taylor summation form
def taylor_gaussian(x,n):
    appx=0
    for i in range(n+1):
        appx += (((-1) ** i) * ((x) ** (2 * i)))/(math.factorial(i))
        
    return appx
````

(d) Calculate the true absolute error between $e^{-x^2}$ evaluated at $x=1.0$ and the Taylor approximation evaluated at $x=1.0$, when $n$ ranges from $[0,20]$ (*i.e.*, up to and including polynomial of order 40).  Plot the true absolute error as a function of $n$ on a semilog y plot (x-axis is linear, y-axis is logarithmic base 10). 

````python
#this is the true value of e^-x^2 evaluated at x=1
one = np.exp(-1**2)

#this function makes a list of all the taylor gaussian appx...
# ... values evalued at a certain x till a certain n
def tay_appr(X,n):
    appx=[]
    for i in range(n+1):
        appx.append(taylor_gaussian(X,i))
    return appx

#this function uses the taylor appx values and...
# ...calculates the true absolute error
def abs_error(approx,true):
    error=[]
    for i in approx:
        error.append(abs(i - true))
    return error
#this calls the two previously defined fucntions
list_error = abs_error(tay_appr(1,20),one)

plt.figure(figsize=(8, 5), dpi=120)

#this is the x axis
X = np.linspace(0, 20, 21, endpoint=True)

#this plots X vs Y, but the semilog(X/Y) makes the plot axis...
# ... half(semi) log, half linear based on the Y or X variable
plt.semilogy(X,list_error)

#plot details
plt.xlabel("n")
plt.ylabel("True absolute error")
plt.title("Plot of the True Absolute Error of $e^{{-x}^2}$ Taylor \
Approxmiation Evaluated at X=1.0")
plt.show()
````
